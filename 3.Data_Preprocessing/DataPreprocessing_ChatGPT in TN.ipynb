```python
dataset.fillna({"District (Where you live in Tamil Nadu)": "Unknown"}, inplace=True)
```

```python
dataset.isna().sum()
```

```python
#Converting Text to Numeric - We can map hours/minutes to a float number of hours.
def convert_time_to_hours(time_str):
    time_str = str(time_str).lower().strip()
    if "half" in time_str:
        return 0.5
    elif "more than 1" in time_str:
        return 1.5
    elif "more than 3" in time_str:
        return 3.5
    elif "1 hour" in time_str:
        return 1
    elif "2 hours" in time_str:
        return 2
    elif "3 hours" in time_str:
        return 3
    else:
        return np.nan

# Apply conversion
dataset["Time Spent Hours"] = dataset["How much spending per day"].apply(convert_time_to_hours)
```

```python
# Check results
print(dataset[["How much spending per day", "Time Spent Hours"]].head())
```

```python
#Convert Numeric Columns - Convert Age, Spending, and Rating columns to numeric safely
dataset["Age"] = pd.to_numeric(dataset["Age"], errors="coerce")
dataset["How much spending per day"] = pd.to_numeric(dataset["How much spending per day"], errors="coerce")
dataset["Rate how helpful ChatGPT is for you (1-5)"] = pd.to_numeric(dataset["Rate how helpful ChatGPT is for you (1-5)"], errors="coerce")
```

```python
dataset.columns
```

```python
dataset.drop('Timestamp', axis=1, inplace=True)
dataset.drop('Name', axis=1, inplace=True)
dataset.drop('How much spending per day', axis=1, inplace=True)
```

```python
dataset.columns = dataset.columns.str.strip()           # removes extra spaces
dataset.columns = dataset.columns.str.replace('\n', ' ')  # replaces line breaks
dataset.columns = dataset.columns.str.replace('\r', ' ')
dataset.rename(columns={
    "District (Where you live in Tamil Nadu)": "District",
    "How often do you use ChatGPT": "Usage Frequency",
    "Main purpose of using ChatGPT": "Main Purpose",
    "Which language do you mainly use ChatGPT in ?": "Language Used",
    "Rate how helpful ChatGPT is for you (1-5)": "User Rating",
    "Did you take actions based on ChatGPT suggestions ?": "Actions Taken",
}, inplace=True)
```

#Encode Categorical Variables
from sklearn.preprocessing import LabelEncoder

categorical_cols = [
    "Gender",
    "District (Where you live in Tamil Nadu)",
    "Occupation",
    "How often do you use ChatGPT",
    "Main purpose of using ChatGPT",
    "Which language do you mainly use ChatGPT in?",
    "Education Level",
    "Did you take actions based on ChatGPT suggestions?",
    "Device Used"
]

label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    dataset[col] = le.fit_transform(dataset[col].astype(str))
    label_encoders[col] = le

#manualy mapping values
dataset['UsageFrequency'] = dataset['UsageFrequency'].map({
    'Rarely': 1,
    'Sometimes': 2,
    'Often': 3,
    'Daily': 4
})

```python
#Ordinal Columns - with order
from sklearn.preprocessing import OrdinalEncoder

ordinal_cols = ['Usage Frequency', 'Education Level']

# define category order explicitly
encoder = OrdinalEncoder(categories=[
    ['Rarely', 'Once a week', '3-5 times a week', 'Daily'],  # order for UsageFrequency
    ['School', 'College', 'Graduate', 'Post Graduate']  # order for EducationLevel
])

dataset[ordinal_cols] = encoder.fit_transform(dataset[ordinal_cols])
```

```python
#Nominal data - Names
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

nominal_cols = ["Gender", "District", "Occupation", "Language Used",
                "Actions Taken", "Device Used"]

# Label Encoding for nominal columns
le_dict = {}
for col in nominal_cols:
    le = LabelEncoder()
    dataset[col] = le.fit_transform(dataset[col].astype(str))
    le_dict[col] = le  # Store encoder to map back if needed
```

```python
# ---------- TARGET COLUMN ----------
target_le = LabelEncoder() 
dataset["Main Purpose"] = target_le.fit_transform(dataset["Main Purpose"].astype(str))
```
